You want to create a new column called Customer Type based on the following conditions:

If tenure is 20 or more, label it as 'good'.
If tenure is between 10 and 19, label it as 'medium'.
If tenure is less than 10, label it as 'bad'.

```python
df["Customer Type"]=np.where(df["tenure"]>=20, "good",
                    np.where(df["tenure"].between(10,19), "medium", "bad"))
df
                    
```
